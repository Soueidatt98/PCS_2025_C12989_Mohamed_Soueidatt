### Ex. 4 – Quick mental check  
`ans = 1 + L[0] + L[-1]` is always **0** whatever `m`.

```python
def ans_val(m):
    L = [n - m / 2 for n in range(m)]
    return 1 + L[0] + L[-1]
[ans_val(k) for k in range(1, 6)]
```

### Ex. 5 – Multistep recursion (placeholder)  

```python
import math, matplotlib.pyplot as plt
h=0.001
a=-0.51
u=[math.exp(0.0),math.exp(a*h),math.exp(2*a*h)]
for _ in range(998):
    u.append(u[-1]+h*a*((23/12)*u[-1]-(4/3)*u[-2]+(5/12)*u[-3]))
td=[n*h for n in range(len(u))]
plt.figure()
plt.plot(td,u)
plt.xlabel('t')
plt.ylabel('U_n')
plt.title('Approximate solution U_n')
plt.figure()
err=[abs(math.exp(a*t)-u_n) for t,u_n in zip(td,u)]
plt.plot(td,err)
plt.xlabel('t')
plt.ylabel('|e^{at}-U_n|')
plt.title('Error')
plt.show()
```

### Ex. 6 – Symmetric difference of two sets

```python
def sym_diff(A, B):
    return (A - B) | (B - A)

A = {1, 2, 3}
B = {3, 4}
sym_diff(A, B), A.symmetric_difference(B)
```

### Ex. 8 – `intersection` vs `intersection_update`  
`intersection` returns a **new** set; `intersection_update` mutates the original.

```python
S = {1, 2, 3}
T = {2, 3, 4}
new_set = S.intersection(T)
mut_set = S.copy()
mut_set.intersection_update(T)
(S, new_set, mut_set)
```
